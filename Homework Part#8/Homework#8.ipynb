## Task#1
https://leetcode.com/problems/find-minimum-in-rotated-sorted-array/

```python
class Solution:
    def additional_func(self, numlist, left, right):
        
        if left == right:
            return numlist[left]

        if numlist[right - 1] > numlist[right]:
            return numlist[right]
            
        elif numlist[left] < numlist[right]:
            return numlist[left]

        mid = int((left + right)/2)

        if numlist[mid] < numlist[mid - 1]:
            return numlist[mid]

        elif numlist[left] > numlist[mid]:
            return self.additional_func(numlist, left, mid - 1)
    
        else: 
            return self.additional_func(numlist, mid + 1, right)
            

    def findMin(self, nums: List[int]) -> int:
        return self.additional_func(nums, 0, len(nums) - 1)
```

## Task#2
https://leetcode.com/problems/search-in-rotated-sorted-array/

```python
class Solution:
    def additional_func(self, numlist, left, right, targ):
        
        if left >= right and numlist[left] != targ:
            return -1

        mid = (left + right)//2

        if numlist[mid] == targ:
            return mid
        elif numlist[left] == targ:
            return left
        elif numlist[right] == targ:
            return right
        
        if numlist[left] > numlist[right]:
            if numlist[mid] > targ and targ < numlist[left] and numlist[mid] < numlist[right]:
                return self.additional_func(numlist, left, mid - 1, targ)
            
            elif numlist[mid] > targ and targ < numlist[left] and numlist[mid] > numlist[right]:
                return self.additional_func(numlist, mid + 1, right, targ)

            elif numlist[mid] < targ and targ > numlist[left] and numlist[mid] < numlist[right]:
                return self.additional_func(numlist, left, mid - 1, targ)
            
            elif numlist[mid] < targ and targ > numlist[left] and numlist[mid] > numlist[right]:
                return self.additional_func(numlist, mid + 1, right, targ)

            elif numlist[mid] > targ and targ > numlist[left]:
                return self.additional_func(numlist, left, mid - 1, targ)
            
            elif numlist[mid] < targ and targ < numlist[left]:
                return self.additional_func(numlist, mid + 1, right, targ)
            

        else:
            if numlist[mid] > targ:
                return self.additional_func(numlist, left, mid - 1, targ)
            else:
                return self.additional_func(numlist, mid + 1, right, targ)


    def search(self, nums: List[int], target: int) -> int:
        return self.additional_func(nums, 0, len(nums) - 1, target)
```

## Task#4
https://leetcode.com/problems/search-in-rotated-sorted-array-ii/

```python
class Solution:
    def additional_func(self, numlist, left, right, targ):
        
        if left >= right and numlist[left] != targ:
            return False

        mid = (left + right)//2

        if numlist[mid] == targ or numlist[left] == targ or numlist[right] == targ:
            return True
        
        if numlist[left] > numlist[right]:
            if numlist[mid] > targ and targ < numlist[left] and numlist[mid] <= numlist[right]:
                return self.additional_func(numlist, left, mid - 1, targ)
            
            elif numlist[mid] > targ and targ < numlist[left] and numlist[mid] > numlist[right]:
                return self.additional_func(numlist, mid + 1, right, targ)
                

            elif numlist[mid] < targ and targ > numlist[left] and numlist[mid] < numlist[right]:
                return self.additional_func(numlist, left, mid - 1, targ)
            
            elif numlist[mid] < targ and targ > numlist[left] and numlist[mid] > numlist[right]:
                return self.additional_func(numlist, mid + 1, right, targ)

            elif numlist[mid] > targ and targ > numlist[left]:
                return self.additional_func(numlist, left, mid - 1, targ)
            
            elif numlist[mid] < targ and targ < numlist[left]:
                return self.additional_func(numlist, mid + 1, right, targ)
            

        elif numlist[left] < numlist[right]:
            if numlist[mid] > targ:
                return self.additional_func(numlist, left, mid - 1, targ)
            else:
                return self.additional_func(numlist, mid + 1, right, targ)
        else:
            return self.additional_func(numlist, left + 1, right - 1, targ)


    def search(self, nums: List[int], target: int) -> bool:
        return self.additional_func(nums, 0, len(nums) - 1, target)



```

## Task#5
https://leetcode.com/problems/maximum-candies-allocated-to-k-children/

```python
class Solution:
    def additional_func(self, left, right, candies, kids, ans):

        if left > right:
            return ans

        mid = (left + right)//2

        sum_candies = 0
        for pile in candies:
            sum_candies += pile//mid

        if sum_candies >= kids:
            if ans < mid:
                return self.additional_func(mid + 1, right, candies, kids, mid)
            else:
                return self.additional_func(mid + 1, right, candies, kids, ans)
        else:
            return self.additional_func(left, mid - 1, candies, kids, ans)


    def maximumCandies(self, candies: List[int], k: int) -> int:
        return self.additional_func(1, max(candies), candies, k, 0)

        
```
